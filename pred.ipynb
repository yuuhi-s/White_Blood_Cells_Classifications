### 出力

```python
images = os.listdir('./out_image')
path = './out_image/'

for image in images:
    if image == '.DS_Store':
        pass
    else:
        img = cv2.imread(path + image)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(15, 15))
        plt.title(image)
        plt.imshow(img);
```
